##PIHM Serial Code Profile

Since PIHM code is written in c, I use GNU profiler which is called “gprof” to profile the code.

####Step 1:
* Compile the code with '-pg' option

```python
%%file makefile
SHELL = /bin/sh

srcdir       = .
builddir     = .
top_builddir = ../../
top_builddir = ../../
prefix       = /dscrhome/xc46/SUNDIALS/sundials
exec_prefix  = ${prefix}
includedir   = /dscrhome/xc46/SUNDIALS/sundials/include
libdir       = /dscrhome/xc46/SUNDIALS/sundials/lib


CPP      = /usr/bin/cc -E
CPPFLAGS =
CC       = /usr/bin/gcc
CFLAGS   = -pg -O0
LDFLAGS  =
LIBS     = -lm
MACRO    = -DSUB_SURF_RIV -DDIFFUSION -DLAYER3
SRC    = pihm.c f.c read_alloc.c initialize.c is_sm_et.c update.c print.c

COMPILER_PREFIX =
LINKER_PREFIX   =

SUNDIALS_INC_DIR = $(includedir)
SUNDIALS_LIB_DIR = $(libdir)
SUNDIALS_LIBS    = -lsundials_cvode -lsundials_nvecserial -lsundials_shared

all:
        @(echo)
        @(echo '       make pihm     - make pihm        ')
        @(echo '       make clean    - remove all executable files')
        @(echo)

pihm:
        @echo '...Compiling PIHM ...'
        @$(CC) $(CFLAGS) -I$(SUNDIALS_INC_DIR) -L$(SUNDIALS_LIB_DIR) -o $(builddir)/pihm $(SRC) $(SUNDIALS_LIBS) $(LIBS) $(MACRO)

clean:
        @rm -f *.o
        @rm -f pihm                        
```

####Step 2: Execute the code
* This will generate a gmon.out file.

####Step 3 : Run the gprof tool to generate the file with profiling information
* $  gprof pihm gmon.out > profile.txt
